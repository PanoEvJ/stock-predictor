How does the Prophet Algorithm differ from an LSTM?
Why does an LSTM have poor performance against ARIMA and Profit for Time Series?

The Prophet algorithm uses statistical modeling to identify trends in the data and make predictions. The Prophet algorithm is applicable for time series forecasting. On the other hand LSTM is not only applied for for time series forecasting but for a broader range of sequentially modeled problems. LSTM is machine learning technique that uses a recurrent neural network architecture that can learn the order dependence between items in a sequence.

There are a few reasons why an LSTM cannot outperform ARIMA and Prophet for Time Series. First, the LSTM requires a much larger amount of data to train effectively. If this amount is not availbe then ARIMA and Prophet will perform better. Second, LSTM's are complex models which makes them prone to overfitting when capturing data fluctuations that simpler model (ARIMA, Prophet) would not. On the contrary, ARIMA and Prophet are preventing overfitting by built-in regularization features. Then, since the LSTM is a more complex model, it is naturally less computationally efficient. LSTM's are also much less interpretable compared to ARIMA and Prophet. 
The LSTM prediction is based on a set of last values, we are therefore less prone to variance due to seasonality and already consider the current trend. In contrast to that, the prophet model is doing a good job modeling as an additive system and finding out and displaying seasonalities.

What is exponential smoothing and why is it used in Time Series Forecasting?

Exponential smoothing is a time series forecasting method. The method is based on the assumption that the most recent observations are the most relevant for forecasting future values, and assigns more weight to these observations than to older ones. In detail, time series methods are based on the principle that a prediction is the weighted linear sum of past observations. The Exponential Smoothing method works by assigning exponentially decreasing weights for past observations. It is called so because the weight assigned to each demand observation is exponentially decreased.

What is stationarity? What is seasonality? Why Is Stationarity Important in Time Series Forecasting?

Seasonality refers to regural & predictable trends and patterns in time series data. This trends repeat at fixed time intervals. Failing to account for seasonality can lead to inaccurate forecasts and poor business decisions.

Stationarity means that the statistical properties of a process generating a time series do not change over time. It does not mean that the series does not change over time, just that the way it changes does not itself change over time. Most simple example of a stationary process is a linear function.

Stationarity is important in Time Series Forecasting because, first of all, it is easier to model and analyze. The simplicity of stationary processes makes them helpful building blocks in constructing approximating more complex phenomena. In many cases involving time series, it is important to determine if the data was generated by a stationary process, and possibly to transform it so it has the properties of a sample generated by such a process.
